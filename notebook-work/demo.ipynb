```python
from pyspark.sql import SparkSession
```

```python
# 1. Initialize Spark Session 
# (Note: If you are using Databricks or Synapse, the 'spark' variable is already created for you)
spark = SparkSession.builder \
    .appName("ReadCSV") \
    .getOrCreate()
```

```python
# 2. Define the path to your CSV file (Local path, S3, ADLS, etc.)
file_path = "../data/bucketing/orders.csv"

# 3. Read the CSV file into a DataFrame
df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)

# 4. Inspect the data
# df.printSchema()   # Displays the column names and data types
df.explain(True)        # Displays the first 5 rows of data
```
